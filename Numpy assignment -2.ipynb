# 1) Write a function, foo_1, that takes a NumPy array arr and returns the indices of the elements that are closest to a given target. If target is not in the given array, it should return ‘no item found’

```python
import numpy as np
```

```python
def foo_1(arr, target):
    if target not in arr:
        return 'no item found'
    else:
        return np.argsort(np.abs(arr - target))

arr = np.array([3, 8, 7, 7, 7, 7, 7, 8, 7, 10])
target = 7

r = foo_1(arr, target)
print(r)
```

```python
target = 11

foo_1(arr, target)
```

np.abs(arr - target): Calculates the absolute differences between each element in the array arr and the target value.

np.argsort(...): Returns the indices that would sort the absolute differences in ascending order. This means it provides the indices of the elements in the array that, when accessed in the sorted order, would give the sorted absolute differences.

if target not in arr:: Checks if the target value is not present in the array. If the target is not in the array, it prints "no item found"

return sol: Returns the sorted indices obtained from np.argsort(...). This array of indices represents the order in which the elements of arr would be sorted based on their absolute differences from the target value.

# 2). Write a user defined function foo_2 that takes a 1 or 2 dim NumPy array arr and returns a new n-1 dim array, where each element is replaced by the without using count of non-zero elements wrt axis.

```python
def foo_2(arr, axis):
    if arr.ndim == 2:
        x=np.where(arr != 0, 1, 0)
        counts=np.sum(x,axis=axis)
        return counts
    elif arr.ndim == 3:
        x=np.where(arr != 0, 1, 0)
        counts=np.sum(x,axis=axis)
        return counts
```

```python
array = np.array([[1, 2, 3, 4],
                [0, 5, 0, 0],
                [2, 0, 1, 0]])
foo_2(array,axis=1)
```

```python
foo_2(array,axis=0)
```

```python
arr1 = np.array([[[1, 2, 3, 4], [0, 5, 0, 0], [2, 0, 1, 0]],
                 [[2, 0, 1, 0], [0, 5, 0, 0], [1, 2, 3, 4]]])
foo_2(arr1,axis=0)
```

if arr.ndim == 2: and elif arr.ndim == 3:: Checks the dimensionality of the input array arr. If it's a 2D array, it performs certain operations, and if it's a 3D array, it performs similar operations.

x = np.where(arr != 0, 1, 0): Creates a binary array (x) where non-zero elements in the input array arr are replaced with 1, and zero elements are replaced with 0.

counts = np.sum(x, axis=axis): Calculates the sum along the specified axis (axis). This effectively counts the number of non-zero elements along the specified axis.

The function then returns the counts array, which represents the count of non-zero elements along the specified axis.

# 3.1)3.1 You are given a 2D NumPy array of data representing student scores. Each row represents a student, and each column represents a subject. Extract the scores of all students who scored above 90 in at least one subject. Provide the resulting array.


```python
second_array=np.array([[85,88,89],  
                      [78,89,94], 
                      [91,87,90],
                      [87,95,84]])
second_array
```

```python
def foo_2(array):
    x=np.any(second_array > 90, axis=1)
    result=array[x]                                   
    print(" students who scored greater than 90:")
    print(result)
```

```python
foo_2(second_array)
```

np.any(array > 90, axis=1): This checks if any element in each row along axis 1 is greater than 90. The result is a boolean array indicating whether the condition is satisfied for each row.

array[x]: This selects the rows where the condition is True (where any element in the row is greater than 90).

print("Students who scored greater than 90:"): This prints a message to indicate what the printed information represents.

print(result): This prints the selected rows, which correspond to students who scored greater than 90.
